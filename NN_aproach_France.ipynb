```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv('data_COVID19.csv')
```

```python
data_frame = data_frame.loc[data_frame['countriesAndTerritories'] == 'France']
```

```python
data_frame = data_frame.loc[::-1,'cases']
data_frame = data_frame.cumsum()
```

```python
y = np.array(data_frame.iloc[:])
```

```python
x = list(range(1, len(y) +1))
```

```python
x  = np.array(x)
# data_len = len(x)
# absss = y[data_len-1]
# y = y / absss
```

```python
import tensorflow as tf
```

```python
NN_france_model = tf.keras.models.Sequential()
```

```python
NN_france_model.add(tf.keras.layers.Dense(50, input_dim=1, activation='relu'))
```

```python
NN_france_model.add(tf.keras.layers.Dense(50, activation='relu'))
```

```python
NN_france_model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
NN_france_model.add(tf.keras.layers.Dense(1, activation='linear'))
```

```python
NN_france_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
NN_france_model.summary()
```

```python
NN_france_model.fit(x, y, epochs=2000, shuffle=False)
```

```python
xd = NN_france_model.predict(x)
```
