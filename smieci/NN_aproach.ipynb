```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv('coronavirus-machine-learning/data/france_history.csv')
```

```python
data_frame =data_frame.loc[data_frame['countriesAndTerritories'] == 'Italy']
```

```python
italy_dict = {"data":data_frame['cases']}
data_italy = pd.DataFrame(data=italy_dict)
```

```python
data_frame=data_frame.reindex(index=data_frame.index[::-1])
```

```python
data_frame.reset_index(drop=True, inplace=True)
```

```python
data_frame['cum_cases'] = data_frame['cases'].cumsum()
```

```python
train_labels = data_frame['cases']
```

```python
import tensorflow as tf
from tensorflow.keras.activations import sigmoid, linear, relu, tanh
```

```python
NN_model = tf.keras.Sequential()
```

```python
NN_model
```

```python
NN_model.add(tf.keras.layers.Dense(50, input_dim=1))
```

```python
NN_model.add(tf.keras.layers.Activation(relu))
```

```python
NN_model.add(tf.keras.layers.Dense(50))
```

```python
NN_model.add(tf.keras.layers.Activation(relu))
```

```python
NN_model.add(tf.keras.layers.Dense(1))
```

```python
NN_model.add(tf.keras.layers.Activation(sigmoid))
```

```python
NN_model.add(tf.keras.layers.Dense(1))
```

```python
NN_model.add(tf.keras.layers.Activation(linear))
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
NN_model.compile(loss='mse', optimizer=optimizer)
```

```python
history = NN_model.fit(train_data, train_labels, epochs=500)
```
